Importando as bibliotecas necessárias:

```python
import pandas as pd
import plotly.express as px
```

Lendo o arquivo .csv:

```python
car_data = pd.read_csv('../vehicles.csv')
```

Visão inicial do DataFrame:

```python
car_data.info()
car_data.sample(5)
```

Criando um histograma para analisar a condição dos carros a partir do ano do modelo:

```python
fig = px.histogram(
    car_data,
    x='model_year',
    color='condition',
    barmode='stack',
    title='Distribuição de condição dos veículos por ano do modelo'
)

fig.update_layout(
    xaxis_title='Ano do Modelo',
    yaxis_title='Quantidade de Veículos',
    legend_title='Condição'
)
```

Criando um gráfico de dispersão:

```python
fig2 = px.scatter(
    car_data,
    x='odometer',
    y='price',
    color='transmission',
    title='Preço VS quilometragem por tipo de câmbio'
)

fig2.update_layout(
    xaxis_title='Quilometragem',
    yaxis_title='Preço',
    legend_title='Tipo de câmbio'
)
```
